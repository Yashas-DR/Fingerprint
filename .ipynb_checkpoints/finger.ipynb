```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt
import os

# Check if CUDA is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# Data transform
data_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Dataset path
data_dir = r'E:\final year project\Blood-Group-Prediction-from-Fingerprints-Using-CNN-main\Code\dataset_blood_group'
image_dataset = datasets.ImageFolder(data_dir, transform=data_transform)

# Split dataset
train_size = int(0.8 * len(image_dataset))
val_size = len(image_dataset) - train_size
train_dataset, val_dataset = random_split(image_dataset, [train_size, val_size])

# Data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4, pin_memory=True)
dataloaders = {'train': train_loader, 'val': val_loader}

# Simple CNN Model
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1 = nn.Linear(64 * 56 * 56, 512)
        self.fc2 = nn.Linear(512, 8)  # 8 classes for blood groups

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  # (32, 224, 224) → (32, 112, 112)
        x = self.pool(F.relu(self.conv2(x)))  # (64, 112, 112) → (64, 56, 56)
        x = x.view(-1, 64 * 56 * 56)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = SimpleCNN().to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# For plotting
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

# Training loop
num_epochs = 25
for epoch in range(num_epochs):
    print(f'\nEpoch {epoch+1}/{num_epochs}')
    
    for phase in ['train', 'val']:
        if phase == 'train':
            model.train()
            print("Training...")
        else:
            model.eval()
            print("Validating...")

        running_loss = 0.0
        running_corrects = 0

        for inputs, labels in dataloaders[phase]:
            inputs = inputs.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()

            with torch.set_grad_enabled(phase == 'train'):
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                loss = criterion(outputs, labels)

                if phase == 'train':
                    loss.backward()
                    optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(dataloaders[phase].dataset)
        epoch_acc = running_corrects.double() / len(dataloaders[phase].dataset)

        print(f'{phase.capitalize()} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        if phase == 'train':
            train_losses.append(epoch_loss)
            train_accuracies.append(epoch_acc.item())
        else:
            val_losses.append(epoch_loss)
            val_accuracies.append(epoch_acc.item())

# Save model
torch.save(model.state_dict(), 'fingerprintbloodgroupmodel.pkl')
print("\nModel saved as fingerprintbloodgroupmodel.pkl")

# Plot Accuracy Graph
plt.figure(figsize=(10, 5))
plt.plot(train_accuracies, label='Train Accuracy', marker='o')
plt.plot(val_accuracies, label='Validation Accuracy', marker='o')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Loss Graph
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Train Loss', marker='x')
plt.plot(val_losses, label='Validation Loss', marker='x')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Create DataFrame
results_df = pd.DataFrame({
    'Epoch': list(range(1, num_epochs + 1)),
    'Train Accuracy': train_accuracies,
    'Validation Accuracy': val_accuracies,
    'Train Loss': train_losses,
    'Validation Loss': val_losses
})

# Format float values to 4 decimal places
pd.set_option('display.float_format', '{:.4f}'.format)

# Print structured table
print("\nTraining Results Table:\n")
print(results_df.to_string(index=False, justify='center'))
```

```python
best_epoch = results_df['Validation Accuracy'].idxmax()
print(f"\nBest Validation Accuracy: {results_df['Validation Accuracy'][best_epoch]:.4f} at Epoch {results_df['Epoch'][best_epoch]}")
```

```python
# Add Accuracy Gap column
results_df['Accuracy Gap'] = results_df['Train Accuracy'] - results_df['Validation Accuracy']

# Plot Accuracy Gap
plt.figure(figsize=(6, 4))
plt.plot(results_df['Epoch'], results_df['Accuracy Gap'], marker='o', linestyle='--', color='purple')
plt.title('Accuracy Gap Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Train - Validation Accuracy')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

class_correct = [0] * 8
class_total = [0] * 8

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        for i in range(len(labels)):
            label = labels[i].item()
            class_correct[label] += (predicted[i] == labels[i]).item()
            class_total[label] += 1

# Bar plot
accuracy_per_class = [100 * c / t if t != 0 else 0 for c, t in zip(class_correct, class_total)]
plt.figure(figsize=(8, 5))
plt.bar(image_dataset.classes, accuracy_per_class, color='teal')
plt.ylabel('Accuracy (%)')
plt.title('Validation Accuracy per Blood Group')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
